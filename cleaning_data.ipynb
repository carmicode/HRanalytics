```python
def preprocess_series(x=None):
  serie = -1
  if type(x) == str:
    x = x.rstrip()
    if x == "Seed":
      serie = x
    elif x == "Pre-Seed":
      serie = x
    else:
      serie = x.split(" ")[1]
  return serie
series = []
for each in df["Stage - Series"]:
  serie = preprocess_series(x=each)
  series.append(serie)
```

```python
df = df.drop(["Stage - Series"], axis=1)
df["stage"] = series
```
